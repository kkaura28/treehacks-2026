```python
import cv2, os

video_path = os.path.abspath(os.path.join(os.getcwd(), "..", "data", "sam3", "surgery_video.mp4"))
cap = cv2.VideoCapture(video_path)

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
duration_s = total_frames / fps if fps > 0 else 0
codec_int = int(cap.get(cv2.CAP_PROP_FOURCC))
codec = "".join([chr((codec_int >> 8 * i) & 0xFF) for i in range(4)])
file_size_mb = os.path.getsize(video_path) / (1024 * 1024)

cap.release()

print(f"Path:       {video_path}")
print(f"Resolution: {width} x {height}")
print(f"FPS:        {fps:.2f}")
print(f"Codec:      {codec}")
print(f"Frames:     {total_frames}")
print(f"Duration:   {duration_s:.2f}s ({duration_s/60:.2f} min)")
print(f"File size:  {file_size_mb:.1f} MB")
print(f"Bitrate:    {file_size_mb * 8 / duration_s:.2f} Mbps" if duration_s > 0 else "")
```
